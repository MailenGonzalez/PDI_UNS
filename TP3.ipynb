*Importamos bibliotecas necesarias:*

```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread
```


*Definimos funciones generales a utilizar:*

```python
#Levanta una imagen dada  
def getImage(path):
  image = imread(path)
  return image

#Retorna los canales r g y b de una imagen  
def getRGB(img):
  r = img[:,:,0]
  g = img[:,:,1]
  b = img[:,:,2]
  return r,g,b

#Construir Imagen a partir de sus componentes RGB
def to_image(r,g,b):
  shape = r.shape
  img = np.zeros((shape[0], shape[1], 3))
  img[:,:,0] = r
  img[:,:,1] = g
  img[:,:,2] = b
  return img.astype(np.uint8)

#Normaliza imagenes entre 0 y 1
def normalize(img):
  return img/255.0

def denormalize(img):
  denorm = np.uint8(img*255)
  return np.clip(denorm,0,255)

#convierte la imagen a una matriz numpy
def imageToArray(img):
  return np.array(img)

#Transforma de RGB a YIQ
def toYIQ(img):
  yiq_matrix = np.array([[0.299, 0.587, 0.114],[ 0.595716, -0.274453, -0.321263],[0.211456, -0.522591, 0.311135]])
  yiq_matrix = np.dot(img, yiq_matrix.T)
  return np.clip(yiq_matrix, a_min=np.array([0,-0.5956,-0.5225]), a_max=np.array([1,0.5956,0.5225]))

#convierte una imagen de YIQ a RGB
def toRGB(img):
  rgb_matrix = np.array([[1, 0.9663, 0.6210],[ 1, -0.2721, -0.6474],[1, -1.1070, 1.7046]])
  rgb_matrix = np.dot(img, rgb_matrix.T)    
  return np.clip(rgb_matrix,0,1)

#grafica una imagen
def plotImage(img):
  plt.imshow(img);

#graficamoslas imagenes
def plotImages(img1, img2, resultImg):
  fig, axes = plt.subplots(1, 3, figsize=(20, 20))
  ax = axes.ravel()
  ax[0].imshow(img1)
  ax[0].set_title("Image 1")
  ax[1].imshow(img2)
  ax[1].set_title("Image 2")
  ax[2].imshow(resultImg)
  ax[2].set_title("Result Image")
  fig.tight_layout()
  plt.show()
```

*Definimos operaciones en RGB*

```python
#suma clampeada en RGB
def clamp_sum_RGB(img1, img2):
  r1, g1, b1 = getRGB(img1)
  r2, g2, b2 = getRGB(img2)
  r_sum = r1 + r2
  g_sum = g1 + g2
  b_sum = b1 + b2
  r_sum = np.clip(r_sum, 0, 255)
  g_sum = np.clip(g_sum, 0, 255)
  b_sum = np.clip(b_sum, 0, 255)
  img_sum = to_image(r_sum,g_sum,b_sum)
  return img_sum

#Resta clampeada RGB
def clamp_sub_RGB(img1, img2):
  r1, g1, b1 = getRGB(img1)
  r2, g2, b2 = getRGB(img2)
  r_sub = r1 - r2
  g_sub = g1 - g2
  b_sub = b1 - b2
  r_sub = np.clip(r_sub, 0, 255)
  g_sub = np.clip(g_sub, 0, 255)
  b_sub = np.clip(b_sub, 0, 255)
  img_sub = to_image(r_sub,g_sub,b_sub)
  return img_sub

#suma promediada RGB
def avg_sum_RGB(img1, img2):
  r1, g1, b1 = getRGB(img1)
  r2, g2, b2 = getRGB(img2)
  r_sum = (r1 + r2)/2
  g_sum = (g1 + g2)/2
  b_sum = (b1 + b2)/2
  r_sum = np.clip(r_sum, 0, 255)
  g_sum = np.clip(g_sum, 0, 255)
  b_sum = np.clip(b_sum, 0, 255)
  img_sum = to_image(r_sum,g_sum,b_sum)
  return img_sum  

#resta promediada RGB
def avg_sub_RGB(img1, img2):
  r1, g1, b1 = getRGB(img1)
  r2, g2, b2 = getRGB(img2)
  r_sub = (r1 - r2)/2
  g_sub = (g1 - g2)/2
  b_sub = (b1 - b2)/2
  r_sub = np.clip(r_sub, 0, 255)
  g_sub = np.clip(g_sub, 0, 255)
  b_sub = np.clip(b_sub, 0, 255)
  img_sub = to_image(r_sub,g_sub,b_sub)
  return img_sub
```

*Definimos operaciones en YIQ:*

```python
#suma clampeada en YIQ
def clamp_sum_YIQ(img1, img2):
  img1_YIQ = toYIQ(normalize(img1))
  img2_YIQ = toYIQ(normalize(img2))
  img_sum = np.copy(img1_YIQ)
  img_sum[:,:,0] = np.clip(img1_YIQ[:,:,0]+ img2_YIQ[:,:,0], 0, 1)
  img_sum[:,:,1] = np.clip((img1_YIQ[:,:,0] * img1_YIQ[:,:,1] + img2_YIQ[:,:,0] * img2_YIQ[:,:,1])/(img1_YIQ[:,:,0]+img2_YIQ[:,:,0]), -0.5956, 0.5956)
  img_sum[:,:,2] = np.clip((img1_YIQ[:,:,0] * img1_YIQ[:,:,2] + img2_YIQ[:,:,0] * img2_YIQ[:,:,2])/(img1_YIQ[:,:,0]+img2_YIQ[:,:,0]), -0.5225, 0.5225)
  img_sum = denormalize(toRGB(img_sum))
  return img_sum

#resta clampeada en YIQ
def clamp_sub_YIQ(img1, img2):
  img1_YIQ = toYIQ(normalize(img1))
  img2_YIQ = toYIQ(normalize(img2))
  img_sum = np.copy(img1_YIQ)
  img_sum[:,:,0] = np.clip(img1_YIQ[:,:,0] - img2_YIQ[:,:,0], 0, 1)
  img_sum[:,:,1] = np.clip((img1_YIQ[:,:,0] * img1_YIQ[:,:,1] - img2_YIQ[:,:,0] * img2_YIQ[:,:,1])/(img1_YIQ[:,:,0]+img2_YIQ[:,:,0]), -0.5956, 0.5956)
  img_sum[:,:,2] = np.clip((img1_YIQ[:,:,0] * img1_YIQ[:,:,2] - img2_YIQ[:,:,0] * img2_YIQ[:,:,2])/(img1_YIQ[:,:,0]+img2_YIQ[:,:,0]), -0.5225, 0.5225)
  img_sum = denormalize(toRGB(img_sum))
  return img_sum

#suma promediada en YIQ
def avg_sum_YIQ(img1, img2):
  img1_YIQ = toYIQ(normalize(img1))
  img2_YIQ = toYIQ(normalize(img2))
  img_sum = np.copy(img1_YIQ)
  img_sum[:,:,0] = np.clip((img1_YIQ[:,:,0]+ img2_YIQ[:,:,0])/2, 0, 1)
  img_sum[:,:,1] = np.clip((img1_YIQ[:,:,0] * img1_YIQ[:,:,1] + img2_YIQ[:,:,0] * img2_YIQ[:,:,1])/(img1_YIQ[:,:,0]+img2_YIQ[:,:,0]), -0.5956, 0.5956)
  img_sum[:,:,2] = np.clip((img1_YIQ[:,:,0] * img1_YIQ[:,:,2] + img2_YIQ[:,:,0] * img2_YIQ[:,:,2])/(img1_YIQ[:,:,0]+img2_YIQ[:,:,0]), -0.5225, 0.5225)
  img_sum = denormalize(toRGB(img_sum))
  return img_sum

#resta promediada en YIQ
def avg_sub_YIQ(img1, img2):
  img1_YIQ = toYIQ(normalize(img1))
  img2_YIQ = toYIQ(normalize(img2))
  img_sum = np.copy(img1_YIQ)
  img_sum[:,:,0] = np.clip((img1_YIQ[:,:,0] - img2_YIQ[:,:,0])/2, None, 1)
  img_sum[:,:,1] = np.clip((img1_YIQ[:,:,0] * img1_YIQ[:,:,1] - img2_YIQ[:,:,0] * img2_YIQ[:,:,1])/(img1_YIQ[:,:,0]+img2_YIQ[:,:,0]), -0.5956, 0.5956)
  img_sum[:,:,2] = np.clip((img1_YIQ[:,:,0] * img1_YIQ[:,:,2] - img2_YIQ[:,:,0] * img2_YIQ[:,:,2])/(img1_YIQ[:,:,0]+img2_YIQ[:,:,0]), -0.5225, 0.5225)
  img_sum = denormalize(toRGB(img_sum))
  return img_sum  

def if_lighter(img1, img2):
  img1_YIQ = toYIQ(normalize(img1))
  img2_YIQ = toYIQ(normalize(img2))
  img_lighter = np.copy(img1_YIQ)
  high, width, channels = img1_YIQ.shape
  for row in range(high):
    for col in range(width):
      Y1= img1_YIQ[row, col, 0]
      Y2= img2_YIQ[row, col, 0]
      if Y1>Y2:
        img_lighter[row,col,0] = Y1
        img_lighter[row,col,1] = img1_YIQ[row, col, 1]
        img_lighter[row,col,2] = img1_YIQ[row, col, 2]
      else:
        img_lighter[row,col,0] = Y2
        img_lighter[row,col,1] = img2_YIQ[row, col, 1]
        img_lighter[row,col,2] = img2_YIQ[row, col, 2]
  return img_lighter

def if_darker(img1, img2):
  img1_YIQ = toYIQ(normalize(img1))
  img2_YIQ = toYIQ(normalize(img2))
  img_darker = np.copy(img1_YIQ)
  high, width, channels = img1_YIQ.shape
  for row in range(high):
    for col in range(width):
      Y1= img1_YIQ[row, col, 0]
      Y2= img2_YIQ[row, col, 0]
      if Y1<Y2:
        img_darker[row,col,0] = Y1
        img_darker[row,col,1] = img1_YIQ[row, col, 1]
        img_darker[row,col,2] = img1_YIQ[row, col, 2]
      else:
        img_darker[row,col,0] = Y2
        img_darker[row,col,1] = img2_YIQ[row, col, 1]
        img_darker[row,col,2] = img2_YIQ[row, col, 2]
  img_darker = denormalize(toRGB(img_darker))      
  return img_darker        
```

## Análisis de Resultados
En el caso de todas las operaciones es necesario controlar que los valores resultantes se encuentren dentro de los rangos (por ejemplo entre 0 y 255 si estamos trabajando en espacio RGB), para no generar resultados erroneos o artefactos en las imagenes. 
Si analizamos cada operacion por separado en el espacio RGB, podemos ver que en el caso de las sumas y restas clampeadas. En este caso, se suman o se restan (depende la operacion) cada uno de los canales de la imagen y se obtiene una nueva imagen, controlando que los valores resultantes esten entre 0 y 255. Se obtienen imagenes muy brillantes y muy oscuras respectivamente, donde la cromaticidad también se ve afectada ya que se obtienen imagenes saturadas. 

En el caso de las operaciones promediadas, las imagenes resultantes no poseen alteraciones en el color, pero si es mas oscura que las imagenes originales debido a que se promedian los valores de las luminancias de entrada.  

En el caso de las operaciones en el espacio YIQ, la luminancia se obtiene como la suma o resta de las luminancias y clampeando en el caso de las operaciones clampeadas o promediando en el caso de las operaciones promediadas. En el caso de la cromaticidad, el color final se ve afectado por la luminancia, es decir que los valores I y Q se forman teniendo en cuenta la luminancia de ambas imagenes, mediante una combinacion convexa de ambas imagenes de entrada. 

En el caso de las operaciones If-Lighter e If-Darker, la imagen resultante se genera con el pixel correspondiente a la imagen de entrada que mayor o menor valor de luminancia posea (dependiendo del tipo de operacion). En este caso, la imagen resultante se forma con pixeles de las imagenes originales, y no se generan nuevos valores.
