Se busca que desarrollen un código que permita abrir una imagen, conocer las características de esta, luego modificarla mediante alguna alteración en los canales RGB, graficarla y luego guardar el resultado como una imagen nueva.

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
#convierte la imagen a un arreglo numpy
def imageToArray(img):
  return np.array(img)
#Abre una imagen dado un path
def openImage(path):
  return Image.open(path)
#Modifica los valores r, g y b en un pixel especifico
def changePixel(img, x, y, r, g, b):
  img[x,y,:] = [r,g,b]
  return img
#Modifica un canal dado en un pixel especifico
def changeOneChannel(img, x, y, channel, value):
  img[x,y, channel] = value
  return img
#genera una imagen PIL desde un arreglo numpy y almacena la imagen.
def saveImage(img_np, path):
  image = Image.fromarray(img_np)
  image.save(path)
#Imprimo y almaceno en un archivo caracteristicas de la imagen
def get_characts(img):
  format = img.format
  size = img.size
  mode = img.mode
  print('Image Format: ', format)
  print('Image Size:', size)
  print('Image Mode:', mode)
  file=open("img_characts",'w')
  file.write('Image Format: ' + str(format)+ '\n')
  file.write('Image Size: ' + str(size)+'\n')
  file.write('Image Mode: ' + str(mode))

def readPixel(img, x, y):
  return img[x,y,:]
```

```python
#Utilizamos una imagen del dataset de skimage, si quisieramos abrir una local, usamos
#la funcion "openImage" que utiliza el método open de PIL
image = openImage("imagen.jpg")
#Imprime las caracteristicas de la imagen
get_characts(image)
#transformamos la imagen a un arreglo numpy
img_array = imageToArray(image)

#visualizamos la imagen 
plt.figure()
plt.title("imagen original")
plt.imshow(img_array)
plt.show()
x =25
y = 25
r= 255
g=255
b=255
#modificamos los 3 canales de un pixel especifico
img_array = changePixel(img_array, x,x,r,g,b)
x= 50
y= 50
#modificamos en un pixel un canal especifico
img_array = changeOneChannel(img_array, x,y, 1, g)

#imprimimos el contenido de un pixel
print("Contenido del pixel [", x,",",y, "] ",readPixel(img_array, 50,50))

#visualizamos la imagen 
plt.figure()
plt.title("Imagen modificada")
plt.imshow(img_array)
plt.show()

#si vemos la imagen almacenada, se ven los pixels que fueron modificados previamente.
saveImage(img_array, 'modifiedImage.jpg')
```
